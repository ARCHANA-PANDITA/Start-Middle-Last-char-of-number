```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({"Id":[101, 102], "mob":[9711387195, 8860571995]})
```

```python
df['mob'] = df['mob'].apply(str)
```

```python
def nyChar(s,x):
    #x=input("From which index you want numbers:")
    for v in df["mob"]:
        if len(s)%2==0:
            if x=="f":
                i = int(len(s))-int(len(s))
                #print(i,s[i],s[i+1])
                return s[i]+s[i+1]
            elif x=="m": 
                i = int(len(s)/2)-1
                #print(i,s[i],s[i+1])
                return s[i]+s[i+1]
            elif x=="l": 
                i = int(len(s))
                #print(i,s[i-2],s[i-1])
                return s[i-2]+s[i-1]
            

    

```

```python
 df['mobile_char'] =df["mob"].apply(nyChar, x="f")
```
